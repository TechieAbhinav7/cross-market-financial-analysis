```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from fredapi import Fred
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
pio.renderers.default= 'notebook'
```

```python
API_Key= '97b06b9373640b144d642f8e5c454816'
```

```python
start_date= '2010-01-01'
end_date= '2025-01-01'
```

```python
fred= Fred(api_key=API_Key)
```

```python
## Real GDP

us_gdp= fred.get_series('NGDPRNSAXDCUSQ', observation_start= start_date, observation_end= end_date).to_frame(name='usa')
germany_gdp= fred.get_series('CPMNACNSAB1GQDE', observation_start= start_date, observation_end= end_date).to_frame(name='germany')
india_gdp= fred.get_series('NGDPRNSAXDCINQ', observation_start= start_date, observation_end= end_date).to_frame(name='india')
```

```python
## Nominal GDP

us_ngdp= fred.get_series('NGDPNSAXDCUSQ', observation_start=start_date, observation_end= end_date).to_frame(name='usa')
germany_ngdp= fred.get_series('NGDPSAXDCDEQ', observation_start=start_date, observation_end= end_date).to_frame(name='germany')
india_ngdp= fred.get_series('NGDPNSAXDCINQ', observation_start=start_date, observation_end= end_date).to_frame(name='india')
```

```python
nominal_gdp= us_ngdp

for i in [germany_ngdp, india_ngdp]:
    nominal_gdp= pd.merge(nominal_gdp, i, left_index=True, right_index=True, how='inner')

real_gdp= us_gdp

for i in [germany_gdp, india_gdp]:
    real_gdp= pd.merge(real_gdp, i, left_index=True, right_index=True, how='inner')
```

```python
real_gdp.index.name= 'date'
nominal_gdp.index.name= 'date'
```
